# Tagging

```python
# B -> begining of entity
# I -> inisde a entity
# O -> no entity found
# eg: 
# New York:  New -> B-geo,  York -> I-geo
# United Nations: United -> B-org,  Nations -> I-org
# at/the/is -> O
```

##### Preprocessing

```python
# we already have one file with enitites assigned to it
# df = pd.read_csv('./tagged_business_cards.csv', encoding='latin1') #encoding='utf8')

with open('./tagged_business_cards.txt', encoding='utf8', errors='ignore') as f:
    text = f.read()
    
data = list(map(lambda x:x.split('\t'), text.split('\n')))
df = pd.DataFrame(data[1:], columns=data[0])


whitespace = string.whitespace
punctuation = "!#$%&\'()*+:;<=>?[\\]^`{|}~"
table_whitespace = str.maketrans('','',whitespace)
table_punctuation = str.maketrans('','',punctuation)

def clean_text(txt):
    txt = str(txt)
    txt = txt.lower()
    remove_whitespace = txt.translate(table_whitespace)
    remove_punctuation = remove_whitespace.translate(table_punctuation)
    
    return str(remove_punctuation)
    
df['text'] = df['text'].apply(clean_text)

# here we will only take the characters required, will remove the mpty shells
cleaned_data = df.query("text != ''")
cleaned_data.dropna(inplace=True)
```

##### convert data into spacy format

```python
group = cleaned_data.groupby('id')
```

```python
all_cards_data = []

for card in group.groups.keys():

    grouparray = group.get_group('000.jpeg')[['text', 'tag']].values
    content = ''
    annotations = {'entities':[]}
    start, end = 0, 0

    for text, label in grouparray:
        text = str(text)
        text_len = len(text) + 1

        start = end 
        end = start + text_len

        if label!='O':
            annot = (start, end-1, label)
            annotations['entities'].append(annot)

        content += text + ' '
        
    all_cards_data.append((content, annotations))
```

```python
# train data preparation
training_data = pickle.load(open('./pickled_data/train_data.pickle', 'rb'))
train_db = DocBin()
for text, annotations in training_data:
    doc = nlp(text)
    ents = []
    for start, end, label in annotations['entities']:
        span = doc.char_span(start, end, label=label)
        ents.append(span)
    doc.ents = ents
    train_db.add(doc)
train_db.to_disk("./train.spacy")
```

```python
# test data preparation
test_data = pickle.load(open('./pickled_data/test_data.pickle', 'rb'))
test_db = DocBin()
for text, annotations in test_data:
    doc = nlp(text)
    ents = []
    for start, end, label in annotations['entities']:
        span = doc.char_span(start, end, label=label)
        ents.append(span)
    doc.ents = ents
    test_db.add(doc)
test_db.to_disk("./test.spacy")
```

```python
# parser
def parser(text, label):
    text = text.lower()
    
    if label=='PHONE':
        text = re.sub(r'\D', '', text)
        
    elif label=='EMAIL':
        allow_special_chars = '@_.\-'
        text = re.sub(r'[^A-Za-z0-9{} ]'.format(allow_special_chars), '', text)
        
    elif label=='WEB':
        allow_special_chars = ':/.%#\-'
        text = re.sub(r'[^A-Za-z0-9{} ]'.format(allow_special_chars), '', text)
        
    elif label in ('NAME', 'DES'):
        text = re.sub(r'[^a-z ]', '', text)
        text = text.title()
        
    elif label=='ORG':
        text = re.sub(r'[^a-z0-9 ]', '', text)
        text = text.title()
        
    return text 

# test parser
parser('Srikanth)&*#$@gmail.com', 'EMAIL')
```

###### Image processing

1. Enhance 
2. GrayScale
3. Blur
4. Edge Detect
5. Morphological transform
6. Countors

```python
def document_scanner(image):

    img_re,size = resizer(image)
    detail = cv2.detailEnhance(img_re,sigma_s = 20, sigma_r = 0.15)
    gray = cv2.cvtColor(detail,cv2.COLOR_BGR2GRAY) # GRAYSCALE IMAGE
    blur = cv2.GaussianBlur(gray,(5,5),0)
    # edge detect
    edge_image = cv2.Canny(blur,75,200)
    # morphological transform
    kernel = np.ones((5,5),np.uint8)
    dilate = cv2.dilate(edge_image,kernel,iterations=1)
    closing = cv2.morphologyEx(dilate,cv2.MORPH_CLOSE,kernel)

    # find the contours
    contours , hire = cv2.findContours(closing,
                                       cv2.RETR_LIST,
                                       cv2.CHAIN_APPROX_SIMPLE)

    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    for contour in contours:
        peri = cv2.arcLength(contour,True)
        approx = cv2.approxPolyDP(contour,0.02*peri, True)

        if len(approx) == 4:
            four_points = np.squeeze(approx)
            break

    cv2.drawContours(img_re,[four_points],-1,(0,255,0),3)

    # find four points for original image
    multiplier = image.shape[1] / size[0]
    four_points_orig = four_points * multiplier
    four_points_orig = four_points_orig.astype(int)

    wrap_image = four_point_transform(image,four_points_orig)
    
    cv2.imshow('original resized to 500', img_re)
    cv2.imshow('detailed', detail)
    cv2.imshow('gray', gray)
    cv2.imshow('blur', blur)
    cv2.imshow('edge', edge_image)
    cv2.imshow('closing', closing)
    cv2.drawContours(img_resized[0], [four_points], -1, (0,255,0), 3)
    cv2.imshow('contour', img_re)
    cv2.imshow('wrap image', wrap_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    return wrap_image, four_points_orig, img_re, closing


img_orig = cv2.imread('./data_with_background/03.jpg')
wrap_image, four_points_orig, img_re, closing = document_scanner(img_orig)
```

```python
# magic color
# it will increase the brightness and contrast of the image
def apply_brightness_contrast(input_img, brightness = 0, contrast = 0):
    
    if brightness != 0:
        if brightness > 0:
            shadow = brightness
            highlight = 255
        else:
            shadow = 0
            highlight = 255 + brightness
        alpha_b = (highlight - shadow)/255
        gamma_b = shadow
        
        buf = cv2.addWeighted(input_img, alpha_b, input_img, 0, gamma_b)
    else:
        buf = input_img.copy()
    
    if contrast != 0:
        f = 131*(contrast + 127)/(127*(131-contrast))
        alpha_c = f
        gamma_c = 127*(1-f)
        
        buf = cv2.addWeighted(buf, alpha_c, buf, 0, gamma_c)

    return buf


magic_image = apply_brightness_contrast(wrap_image, brightness = 30, contrast = 30)
cv2.imshow('wrap image', wrap_image)
cv2.imshow('magic image', magic_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
